```python
# add distance
d = gsw.distance(ctdsection.longitude.values, ctdsection.latitude.values)/1e3  # distance between each station, km
dist = np.cumsum(np.hstack((0,d))) #cumulative sum of dist

ctdsection = ctdsection.assign_coords({'distance': xr.DataArray(dist, coords = [ctdsection.stn], dims = 'stn')
                                      })
ctdsection.distance.attrs['units'] = 'km'
```

## Buoyancy frequency, $N^2$

```python
# mean density across the transect
ctd = ctdsection.mean(dim='stn')

# buoyancy frequency
N2 = p_mid = [0]*(len(ctd.pressure))
[N2, p_mid] = gsw.stability.Nsquared(ctd.salinity, ctd.temperature, ctd.pressure, ctdsection.latitude.mean())

f = gsw.f(ctdsection.latitude.mean())
np.sqrt(N2.mean())/f.values
```

## CMEMS SLA

```python
# Load 2017 global CMEMS data
da = xr.open_mfdataset('../Data/CMEMS/*.nc',parallel=True)
```

```python
# For convinience, change longitude from [0,360] to [-180,180]
with xr.set_options(keep_attrs=True):
    da = da.assign({'longitude':(((da.longitude + 180) % 360) - 180)})
```

```python
# Select data within Mengyang's regional box
BOX = [-160,-156,23.5,28.5]

# rename it to avoid loading the dataset everytime
satellite = da.where((da.longitude>BOX[0])&(da.longitude<BOX[1])&\
              (da.latitude>BOX[2])&(da.latitude<BOX[3]),
               drop=True)
```

```python
# select the another subset to match latitude range of ADCP
satellite = satellite.where((satellite.latitude>=min(adcp300.lat.values))&
                            (satellite.latitude<=max(adcp300.lat.values)), drop=True)
satellite = satellite.sel(time=slice('2017-06-29','2017-07-01'))

# take mean over time
sla = satellite.sla.mean(dim={'time'})
```

```python
# Interpolate SLA data to ADCP track
loni, lati = np.meshgrid(satellite.longitude,satellite.latitude)

# get southbound ADCP track 
lonadcp = adcp300.lon[313::5]
latadcp = adcp300.lat[313::5]

# generate interpolator
f = interp2d(loni, 
             lati,
             sla,
             kind='linear'
            )

# interpolate 
slai = [f(x,y)[0] for x,y in zip(lonadcp,latadcp)]
slai = np.array(slai)

# sla_int = xr.DataArray(
#     sla_f,
#     dims = ('longitude','latitude'),
#     coords = { 'longitude': adcp300.lon.values,
#                'latitude': adcp300.lat.values})
```

## Steric height

```python
# Calculate steric height 
nlat, ndepth = ctdsection.salinity.shape

dyn_height = np.empty((nlat,ndepth))

sal  = ctdsection.salinity.values
temp = ctdsection.temperature.values

sal[np.isnan(sal)] = 0 
temp[np.isnan(temp)] = 0 


for i in range(nlat):
    dyn_height[i] = gsw.geo_strf_dyn_height(
         SA=sal[i], 
         CT=temp[i],
         p=ctdsection.pressure,
         p_ref=450
    )
    
ssh = dyn_height[:,0]/9.81   
```

```python
# now plot both 300-Hz and 38-Hz data
# adcp300_subset = adcp300.where(adcp300.depth<=80,drop=True)
# adcp38_subset = adcp38.where(adcp38.depth>=60,drop=True)
```

```python
# now plot both 300-Hz and 38-Hz data
xlim = (24.4, 28)

fig = plt.figure(figsize=(8.5,8))

gs = fig.add_gridspec(ncols=1, nrows=6)

ax = fig.add_subplot(gs[0])

plt.plot(ctdsection.latitude, ssh-np.nanmean(ssh),
         color='k', linewidth=3
        )

# sla relative to the mean sla over the domain
plt.plot(latadcp, slai-slai.mean(),
         color='r',linewidth=3
        )

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_position(('axes', 1.15))

plt.xlim(*xlim)
plt.ylim(-.25,.25)
plt.xticks([])

plt.ylabel(r'[m]')

ax.annotate("Steric height anomaly",
            xy=(0.2, 0.35), xycoords='axes fraction',
            xytext=(0.1, .85), textcoords='axes fraction',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )

ax.annotate("Sea level anomaly",
            xy=(0.85, 0.6), xycoords='axes fraction',
            xytext=(0.75, .99), textcoords='axes fraction',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3",
                            color='r'),
            color='r'
            )

ax = fig.add_subplot(gs[1:])

kw = dict(levels = np.arange(-0.7,0.7,0.1),add_colorbar=False)

adcp300.u.plot.contourf(x='lat',y='depth',**kw)
adcp38.u.plot.contourf(x='lat',y='depth',**kw)

ctdsection.density.plot.contour('latitude','pressure', levels = 20)

ax.plot(ctdsection.latitude, 
        ctdsection.distance.size*[0],
        'kv', markersize=15)

plt.ylim(500,0)
plt.xlim(*xlim)

plt.savefig('figures/adcp_isopycnal_downcast',
           dpi=200,
           bbox_inches='tight',
           transparent=True,
           )
```
