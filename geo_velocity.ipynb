```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
from matplotlib import cm
import os
import gsw
```

```python
# Load 2017 global CMEMS data
da = xr.open_mfdataset('../Data/CMEMS/all/*.nc',parallel=True)
```

```python
da
```

```python
# For convinience, change longitude from [0,360] to [-180,180]
with xr.set_options(keep_attrs=True):
    da = da.assign({'longitude':(((da.longitude + 180) % 360) - 180)})
```

```python
# Select data within Mengyang's regional box
BOX = [-160,-156,23.5,28.5]
#BOX = [-90,-40,20,50]
#BOX = [110,170,20,50]

# name the subset as ds, to avoid loading the dataset everytime
ds = da.where((da.longitude>BOX[0])&(da.longitude<BOX[1])&\
              (da.latitude>BOX[2])&(da.latitude<BOX[3]),
               drop=True)
```

```python
# velocity u
kw = {'levels': np.arange(-.5,0.5,0.05),
      'extend': 'both'}
ds.isel(time=180).ugos.plot.contourf('longitude','latitude',**kw)
```

```python
# velocity v
kw = {'levels': np.arange(-.5,0.5,0.05),
      'extend': 'both'}
ds.isel(time=180).vgos.plot.contourf('longitude','latitude',**kw)
```

```python
#velosity field at a snapshot

t =180;

lon = ds.isel(time=t).longitude
lat = ds.isel(time=t).latitude

x,y = (lon,lat)
u = ds.isel(time=t).ugosa
v = ds.isel(time=t).vgosa

dist1 = gsw.distance([-160,-156],[23.5,23.5])
dist2 = gsw.distance([-156,-156],[23.5,28.5])

# sla
kw = {'levels': np.arange(-0.28,0.28,0.04),
      'extend': 'both',
      'figsize': (dist1[0]/1e5,dist2[0]/1e5)}

c = ds.isel(time=180).sla.plot.contourf('longitude','latitude',**kw)

# If you want to modify the colorbar/labels on your own
#c = ds.isel(time=180).sla.plot.contourf('longitude','latitude',**kw,add_colorbar=False,add_labels=False)
#cb = plt.colorbar(c,orientation = 'vertical')
#cb.set_label(label = 'Sea level anamoly [m]',size = 15)
#cb.ax.tick_params(labelsize=12)

#c = plt.contourf(lon,lat,ds.isel(time=t).sla,**kw,cmap=cm.RdBu_r)
#plt.colorbar(c)

q = plt.quiver(x,y,u,v)
plt.quiverkey(q, 0.8, 0.9, U = 0.5, label = '0.5m/s',labelpos = 'E',
             coordinates = 'figure')

fig = plt.gcf()
#fig.set_size_inches(dist1[0]/1e5*2, dist2[0]/1e5*2)
fig.savefig('sla_vel.png', dpi=200)

# need to adjust the length of the figure, according to the distance
```

```python
# Load ADCP data
adcp300 = xr.open_dataset('../ADCP_analysis/proc/wh300/contour/wh300.nc')
#adcp38bb = xr.open_dataset('../ADCP_analysis/proc/os38bb/contour/os38bb.nc')
#adcp38nb = xr.open_dataset('../ADCP_analysis/proc/os38nb/contour/os38nb.nc')
```

```python
adcp300 = adcp300.set_coords({'lon','lat','depth'}, inplace=None)
#adcp38nb = adcp38nb.set_coords({'lon','lat','depth'}, inplace=None)
```

```python
adcp = adcp300.sel(time=slice('2017-06-29','2017-07-01'))
```

```python
adcp_subset = adcp.isel(depth_cell=0) # velocity ar surface, 11.1m
```

```python
#velosity field at a snapshot

t =180;

dist1 = gsw.distance([-160,-156],[23.5,23.5])
dist2 = gsw.distance([-156,-156],[23.5,28.5])

# sla
kw = {'levels': np.arange(-0.28,0.28,0.04),
      'extend': 'both',
      'figsize': (dist1[0]/1e5,dist2[0]/1e5)}

c = ds.isel(time=180).sla.plot.contourf('longitude','latitude',**kw)

# If you want to modify the colorbar/labels on your own
#c = ds.isel(time=180).sla.plot.contourf('longitude','latitude',**kw,add_colorbar=False,add_labels=False)
#cb = plt.colorbar(c,orientation = 'vertical')
#cb.set_label(label = 'Sea level anamoly [m]',size = 15)
#cb.ax.tick_params(labelsize=12)

#c = plt.contourf(lon,lat,ds.isel(time=t).sla,**kw,cmap=cm.RdBu_r)
#plt.colorbar(c)

lon = adcp_subset.lon
lat = adcp_subset.lat
x,y = (lon,lat)

u = adcp_subset.u
v = adcp_subset.v

dec = 7
q = plt.quiver(x[::dec],
               y[::dec],
               u[::dec],
               v[::dec],
               scale=5)
plt.quiverkey(q, 0.8, 0.9, U = 0.5, label = '0.5m/s',labelpos = 'E',
             coordinates = 'figure')

fig = plt.gcf()
fig.savefig('sla_adcp.png', dpi=200)
```
